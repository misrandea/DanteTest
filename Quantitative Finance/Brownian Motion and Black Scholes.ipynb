```python
# import libraries:
import numpy as np
import math
from scipy.stats import norm
```

```python
print('Project 1: convergence of a walk on a discrete lattice')

def walk_sim(N_sim, N, mu, T, sigma, X_init):
   # N_sim - number of simulations
   # N - number of timesteps
   # mu - drift
   # T - expiry time
   # sigma - volatility
   # X_init - initial value
   delt = T/N
   up = sigma*math.sqrt(delt)
   down = -sigma*math.sqrt(delt)
   p = 1./2.*(1. + mu/sigma*math.sqrt(delt))
   X_new = np.ones(N_sim)*X_init
   ptest = np.zeros(N_sim)
   for i in np.arange(N): #timestep loop
      # now for each timestep, generate info
      # for all simulations
      ptest = np.random.uniform(0,1,N_sim)
      ptest = (ptest <= p)
      X_new = X_new + ptest*up + (1-ptest)*down
      # note: asterisks here perform elementwise
      # multiplication on numpy vectors, not
      # a dot product.
   # end of generation of data for all simulations
   # for this timestep
   return X_new


T = 2.0
sigma = 0.4
alpha = 0.25
X_init = 3
M = [1000, 10000, 100000]
N = 100

# anticipated values
print('for comparison, the anticipated exact mean is', X_init + alpha*T, 
      ';and the anticipated standard deviation is', np.sqrt(np.square(sigma)*T))



import matplotlib.pyplot as plt
from scipy.stats import norm




# 1000 simulation
print('1)')
print('number of simulation is:',M[0])
lattice1 = walk_sim(M[0], N, alpha, T, sigma, X_init)
lattice1
print('mean of {X_N} =', np.mean(lattice1))
print('standard deviation of {X_N} =', np.std(lattice1))
plt.hist(lattice1, bins=N+1, density=True, alpha=0.6)
plt.title('Histogram of the observed probablity density')

mu, std = norm.fit(lattice1)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
p = norm.pdf(x, mu, std)  
plt.plot(x, p*2.5)

plt.show()

# 10000 simulation
print('2)')
print('number of simulation is:',M[1])
lattice2 = walk_sim(M[1], N, alpha, T, sigma, X_init)
lattice2   
print('mean of {X_N} =', np.mean(lattice2))
print('standard deviation of {X_N} =', np.std(lattice2))
plt.hist(lattice2, bins=N+1,density=True, alpha=0.6)
plt.title('Histogram of the observed probablity density')

mu, std = norm.fit(lattice2)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
p = norm.pdf(x, mu, std)  
plt.plot(x, p*2.5)

plt.show()

# 100000 simulation
print('3)')
print('number of simulation is:',M[0])
lattice3 = walk_sim(M[2], N, alpha, T, sigma, X_init)
lattice3   
print('mean of {X_N} =', np.mean(lattice3))
print('standard deviation of {X_N} =', np.std(lattice3))
plt.hist(lattice3, bins=N+1,density=True, alpha=0.6)
plt.title('Histogram of the observed probablity density')

mu, std = norm.fit(lattice3)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
p = norm.pdf(x, mu, std)  
plt.plot(x, p*2.5)

plt.show()  

print('I observed that with the number of simulation increases, the histogram is shaped more as a histogram.')

```

```python
print("Project 2: Brownian motion and Ito calculus")

import numpy as np
import math
np.random.seed(2022)

N_sim = 100000
N = np.array([100,200,400,800,1600])
T = 2
dt = T/N


# N = 100
dz = np.zeros(N_sim)
zi = dz
I1 = zi*dz
I2 = zi
for i in np.arange(N[0]): #timestep loop
    # now for each timestep, generate info
    # for all simulations
    phi = np.random.normal(0,1,N_sim)
    dz = phi*math.sqrt(dt[0]) # generate dz_i
    I1 = I1+zi*dz
    zi=zi+dz # z_i updated to z_i+1
    I2 = I2+zi*dz
    
print('   1. N = 100')
print('      the expected value of I1 is', np.mean(I1)) # expected value of I1
print('      the expected value of I2 is', np.mean(I2)) # expected value of I1


# N = 200
dz = np.zeros(N_sim)
zi = dz
I1 = zi*dz
I2 = zi
for i in np.arange(N[1]): #timestep loop
    # now for each timestep, generate info
    # for all simulations
    phi = np.random.normal(0,1,N_sim)
    dz = phi*math.sqrt(dt[1]) # generate dz_i
    I1 = I1+zi*dz
    zi=zi+dz # z_i updated to z_i+1
    I2 = I2+zi*dz
    
print('   2. N = 200')
print('      the expected value of I1 is', np.mean(I1)) # expected value of I1
print('      the expected value of I2 is', np.mean(I2)) # expected value of I1

# N = 400
dz = np.zeros(N_sim)
zi = dz
I1 = zi*dz
I2 = zi
for i in np.arange(N[2]): #timestep loop
    # now for each timestep, generate info
    # for all simulations
    phi = np.random.normal(0,1,N_sim)
    dz = phi*math.sqrt(dt[2]) # generate dz_i
    I1 = I1+zi*dz
    zi=zi+dz # z_i updated to z_i+1
    I2 = I2+zi*dz
    
print('   3. N = 400')
print('      the expected value of I1 is', np.mean(I1)) # expected value of I1
print('      the expected value of I2 is', np.mean(I2)) # expected value of I1


# N = 800
dz = np.zeros(N_sim)
zi = dz
I1 = zi*dz
I2 = zi
for i in np.arange(N[3]): #timestep loop
    # now for each timestep, generate info
    # for all simulations
    phi = np.random.normal(0,1,N_sim)
    dz = phi*math.sqrt(dt[3]) # generate dz_i
    I1 = I1+zi*dz
    zi=zi+dz # z_i updated to z_i+1
    I2 = I2+zi*dz
    
print('   4. N = 800')
print('      the expected value of I1 is', np.mean(I1)) # expected value of I1
print('      the expected value of I2 is', np.mean(I2)) # expected value of I1


# N = 1600
dz = np.zeros(N_sim)
zi = dz
I1 = zi*dz
I2 = zi
for i in np.arange(N[4]): #timestep loop
    # now for each timestep, generate info
    # for all simulations
    phi = np.random.normal(0,1,N_sim)
    dz = phi*math.sqrt(dt[4]) # generate dz_i
    I1 = I1+zi*dz
    zi=zi+dz # z_i updated to z_i+1
    I2 = I2+zi*dz
    
print('   5. N = 1600')
print('      the expected value of I1 is', np.mean(I1)) # expected value of I1
print('      the expected value of I2 is', np.mean(I2)) # expected value of I1

print('To conclude:')
print('N |    100      200        400      800      1600')
print('I1|   0.0006   0.0030   -0.0060   -0.0006   -0.0072')
print('I2|   2.0017   2.0036   1.99345   1.99955   1.9926')

print('We observed that the simulation results are consistent with what we expect, that I1(T, delta_t) is approximately 0, and I2(T, delta_t) roughtly equals to T=2. This is a good representatio of Ito\'s property, that each time interval is independent of others.')

```

```python
print("Project 3: Black-Scholes Option Prices and Deltas")

def blsprice(Price, Strike, Rate, Time, Volatility):
    # Function calcutes the Black-Scholes price of call and put option on the underlying stock expires at time T on strike K
    
    sigma_sqrtT = Volatility * np.sqrt (Time)

    d1 = 1 / sigma_sqrtT * (np.log(Price / Strike) + (Rate + Volatility**2 / 2) * Time)
    d2 = d1 - sigma_sqrtT

    phi1 = norm.cdf(d1)
    phi2 = norm.cdf(d2)
    disc = np.exp (-Rate * Time)
    F    = Price * np.exp ((Rate) * Time)

    Call = disc * (F * phi1 - Strike * phi2)
    Put  = disc * (Strike * (1 - phi2) + F * (phi1 - 1))
    return Call, Put


def blsdelta(Price, Strike, Rate, Time, Volatility):
    # Function calcutes the Black-Scholes delta of call and put option on the underlying stock expires at time T on strike K
    
    d1 = 1 / (Volatility * np.sqrt(Time)) * (np.log (Price / Strike) + (Rate + Volatility**2 / 2) * Time)

    phi = norm.cdf(d1)

    CallDelta = phi
    PutDelta = phi - 1
    return CallDelta, PutDelta

print('At-the-money option of a stock of current price $100 with drift rate 0.03 and volatility of 0.25:')

print('Price of (call, put) is',blsprice(100,100,0.03,1,0.25))
print('Delta of (call, put) is',blsdelta(100,100,0.03,1,0.25))
```
